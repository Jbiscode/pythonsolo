# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 윈도우시
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # 맥일시
matplotlib.rcParams['font.size'] = 15                   # 폰트 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False       # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungso-Bold','size':30})
```

```python
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x,y)
plt.xlabel('X축',color = 'red')
plt.ylabel('Y축', color = 'blue')
```

```python
plt.plot(x,y)
plt.xlabel('X축',color = 'red', loc = 'right')     # left,center, right
plt.ylabel('Y축', color = 'blue', loc = 'top')   # top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([2,4,6,8,10])
plt.show()
```
